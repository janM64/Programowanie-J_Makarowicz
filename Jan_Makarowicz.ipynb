```python
docs = list(nlp.pipe(df_kopia["text"].astype(str)))
```

```python
for doc in docs[:5]:  #3.1
  print(doc)
```

```python
for doc in docs:
  for ent in doc.ents:
    if ent.label_ == "persName":
      print(ent.label, ent.text)
```

```python
names_list = []
for doc in docs:
  names = [ent.text for ent in doc.ents if ent.label_ == "persName"]
  names_list.append(", ".join(names))
```

```python
df_kopia["person"] = names_list
```

```python
df_kopia["person"]
```

```python
for doc in docs: #3.2
  for ent in doc.ents:
    if ent.label_ == "placeName":
      print(ent.label, ent.text)
```

```python
place_list = []
for doc in docs:
  place = [ent.text for ent in doc.ents if ent.label_ == "placeName"]
  place_list.append(place)
```

```python
df_kopia["places"] = place_list
```

```python
df_kopia["places"]
```

```python
for doc in docs: #3.3
  for ent in doc.ents:
    if ent.label_ == "orgName":
      print(ent.label, ent.text)
```
